# Import libraries
Seaborn automatically installs these libraries

- *numpy*
- *scipy*
- *pandas*
- *Matplotlib*

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
phool= sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x= "sepal_length", y="sepal_width", data=phool)

#show plot
plt.show()
```

## Adding titles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
phool= sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x= "sepal_length", y="sepal_width", data=phool)

#adding tilte
plt.title("Phoolon ka plot")

#show plot
plt.show()
```

## Adding limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
phool= sns.load_dataset("iris")

#draw a line plot
sns.lineplot(x= "sepal_length", y="sepal_width", data=phool)

#adding title
plt.title("Phoolon ka plot")

#adding limits
plt.xlim(2)
plt.ylim(2)

#show plots
plt.show()
```
